```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(1,10)
print(f"x = {x}")


def f(k,x,m):
    y = k*x + m
    return y
# y= 2x +2
print(f" y = {f(2,x,2)}")



plt.plot(x,f(1,x,1))
plt.title(" EN rät linije")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0,10)

def f(k,x,m):
    y = k*x + m
    return y
    
        


plt.plot(x,f(3,x,1))
plt.xlabel("x")
plt.ylabel("y")
plt.title("funktionen av 3x+1")
plt.show()


```
